```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset (if not loaded)
df = pd.read_csv('../data/customer-churn.csv')  # path adjust করা লাগতে পারে

# Convert 'Date' column to datetime
df['Date'] = pd.to_datetime(df['Date'])

# Aggregate monthly sales
monthly_sales = df.groupby(df['Date'].dt.to_period('M'))['Sales'].sum()
monthly_sales.plot(kind='line', figsize=(12,6), title='Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.show()
```

```python
print(df.columns)
```

```python
# Convert TotalCharges to numeric
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')

# Missing values fill করা
df['TotalCharges'].fillna(df['TotalCharges'].median(), inplace=True)
```

```python
print(df.isnull().sum())
```

```python
print(df.dtypes)
```

```python
# Churn কে 1 এবং 0 এ convert করা
df['Churn'] = df['Churn'].map({'Yes': 1, 'No': 0})
```

```python
# categorical columns identify করা
categorical_cols = df.select_dtypes(include=['object']).columns.tolist()
print(categorical_cols)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# উদাহরণ: Gender distribution
sns.countplot(x='gender', data=df)
plt.show()

# Churn vs Gender
sns.countplot(x='Churn', hue='gender', data=df)
plt.show()
```
